## About numpy save and load

```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.random.rand(2,3)
```

```python
np.save('arr1.npy',arr1)
np.save('arr2.npy',arr2)
```

```python
print(np.load('arr1.npy'))
print(np.load('arr2.npy'))
```

## With numpy logo

```python
loaded=np.load("numpy-logo.npy")
plt.figure(figsize=(10,5),dpi=100)
plt.subplot(121)
plt.imshow(loaded)
plt.title("Numpy logo")
plt.show()

darkloaded=1-loaded
plt.figure(figsize=(10,5),dpi=100)
plt.subplot(122)
plt.imshow(darkloaded)
plt.title("Negative Numpy logo")
plt.show()


```
