```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])
print(sales_data.shape)
print("Sample data for 1st 3 restau: ",sales_data[:3])
print("Sample data for last 2 restau: ",sales_data[-2:])
print(sales_data[:,1:])
```

## Total sales per year

```python
print("total_sales_in 4 year: ",np.sum(sales_data[:,1:],axis=0))
print(np.sum(sales_data[:,1:],axis=1))
```

## Minimum sales per restaurant


```python
print("Minimum sales of every restaurant: ",np.min(sales_data[:,1:],axis=1))
```

## Maximum sales per year

```python
print("Maximum sales per year: ",np.max(sales_data[:,1:],axis=0))
```

## Average sales per restaurant

```python
avg_sales=np.mean(sales_data[:,1:],axis=1)
print("Average sales of every restaurant: ",avg_sales)
```

### Cumulative sales

```python
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print("Per restaurant: ",cumsum)
plt.figure(figsize=(8, 6))
plt.title("Average cumulative sales accross all restaurant ")
plt.plot(np.mean(cumsum,axis=0),'r-o')
plt.grid(True)
plt.xlabel("Year")
plt.ylabel("Average sales")
plt.show()
```

## Vector related

```python
vector1=np.array([1,2,3,4,5])
vector2=np.array([6,7,8,9,10])
print("Sum : ",vector1+vector2)
print("Multiplication : ",vector1*vector2)
print("Dot product : ",np.dot(vector1,vector2))
angle=np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print("Angle between two vectors: ",angle)
```

## Vectorized

```python
restaurant_types = np.array(['biryani', 'chinese', 'pizza', 'burger', 'cafe'])
vectorized_upper=np.vectorize(str.upper)
print(vectorized_upper(restaurant_types))
vectorized_title=np.vectorize(str.title)
print(vectorized_title(restaurant_types))
```
